```python
## using Ordeinal encoding to encode the age

def age_encoding(Student_Age):
    if Student_Age =='18':
        return 1
    elif Student_Age=='19-22':
        return 2
    elif Student_Age=='23-27':
        return 3
    else:
        return None
```

```python
df['Age'] = df['Student_Age'].apply(age_encoding)
```
